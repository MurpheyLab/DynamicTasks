```python
import numpy as np
import scipy
from scipy import signal
import math
%matplotlib inline 
import matplotlib.pyplot as plt

from cartpole import Cartpole
```

```python
#define model
dt = 0.04

cartpole_model = Cartpole(1.0, 0.1, 0.5, 9.8, dt)
```

```python
#define parameters and initial state

T = 40
n = 41
t_span = np.linspace(0,T,n)

state = np.array([0,0,0.1,0])
state0 = state.copy()

u_arr = np.zeros((1,n))

target = np.zeros((4, n))
target[0,:] = 0.2


Q = np.diag([1000.0, 1, 100, 1])
R = np.diag([1.0])

x0_test = np.zeros((4, n))
xu0 = np.concatenate((x0_test, u_arr), axis=0)

def loss(xu):

    xu = xu.reshape((5,n))
    x = xu[0:4, :]
    u = xu[4:5, :]
    
    #single step cost
    result_sum = 0.0
    for step in range(0, x.shape[1]):
        single_target = target[:, step]
        single_x = x[:, step]
        
        single_diff = np.array([single_target - single_x])
        
        single_cost = single_diff@Q@single_diff.T
        
        result_sum += single_cost[0][0]
        
        
    
    return result_sum#np.sum(np.square(target-y))


sigma = np.diag([1000, 1, 100, 1])
inv_sigma = np.linalg.inv(sigma)##
p_cov = 0.01*np.eye(4)
p_target = scipy.stats.multivariate_normal(mean=[0.2, 0.0, 0.0, 0.0], cov=p_cov)
R = 0.1

def ergodic_loss(xu):
    xu = xu.reshape((5,n))
    x = xu[0:4, :]
    u = xu[4:5, :]
    
    
    result_sum = 0.0
    
    #s_sample = np.random.uniform(-3, 3, size=(50,4))
    s_sample = p_target.rvs(size=20) ##sampling in a boundary

    sample_pdf = p_target.pdf(s_sample)
    for si in range(0, len(s_sample)):
        psi = sample_pdf[si]
        
        q_sum = 0.0
        for step in range(0, x.shape[1]):
            sx = np.array([s_sample[si] - x[:,step]]) #only consider x and theta
            norm = sx@inv_sigma@sx.T
            
            q_sum += np.exp(-0.5*norm[0][0])*0.04 #*dt
            
        result_sum += psi*np.log(q_sum)
        
    result_sum = -result_sum
    for step in range(0,x.shape[1]):
        result_sum += R*u[:,step][0]**2
        

    return result_sum
            
            
            
#print(ergodic_loss(0.1*u_arr, [0,0,0.1,0]))



def dynamics_constraint(xu):
    xu = xu.reshape((5,n))
    x = xu[0:4, :]
    u = xu[4:5, :]
    error = np.zeros((1,4*n))
    j = 0
    for i in range(n-1):
        curr_x = x[:,i]
        curr_u = u[:,i]
        obs = np.concatenate((curr_x, curr_u))
        state = cartpole_model.step(obs)
        
        
        error[:,j] = x[0, i+1] - state[0]
        error[:,j+1] = x[1, i+1] - state[1]
        error[:,j+2] = x[2, i+1] - state[2]
        error[:,j+3] = x[3, i+1] - state[3]
        j = j + 4
        
    error[:,j] = x[0,0] - 0.0
    error[:,j+1] = x[1,0] - 0.0
    error[:,j+2] = x[2,0] - 0.1
    error[:,j+3] = x[3,0] - 0.0
    
        
    return error.flatten()

#print(dynamics_constraint(xu0))
# state_arr = prediction(u_arr, t_span, state0, cartpole_model)

# plt.figure()
# plt.subplot(4,1,1)
# plt.plot(t_span, state_arr[:,0] ,'r-',linewidth=2)
# plt.ylabel('x')
# plt.subplot(4,1,2)
# plt.plot(t_span, state_arr[:,1] ,'b-',linewidth=2)
# plt.ylabel('dx')
# plt.subplot(4,1,3)
# plt.plot(t_span, state_arr[:,2] ,'g-',linewidth=2)
# plt.ylabel('theta')
# plt.subplot(4,1,4)
# plt.plot(t_span, state_arr[:,3] ,'m-',linewidth=2)
# plt.ylabel('dtheta')
# plt.show()
```

```python
con = ({'type': 'eq', 'fun': dynamics_constraint})
print(xu0.shape)
result = scipy.optimize.minimize(ergodic_loss,xu0, bounds=[(-3,3),(-3,3),(-3,3),(-3,3),(-1.0, 1.0)],constraints=con)
print(result)
policy = result.x.reshape((5,n))

print(policy.shape)
#print(policy)
```

```python
#state_arr = prediction(policy, t_span, state0, cartpole_model)
plt.figure()
plt.subplot(5,1,1)
plt.plot(t_span, policy[0,:] ,'r-',linewidth=2)
plt.ylabel('x')
plt.subplot(5,1,2)
plt.plot(t_span, policy[1,:] ,'b-',linewidth=2)
plt.ylabel('dx')
plt.subplot(5,1,3)
plt.plot(t_span, policy[2,:] ,'g-',linewidth=2)
plt.ylabel('theta')
plt.subplot(5,1,4)
plt.plot(t_span, policy[3,:] ,'m-',linewidth=2)
plt.ylabel('dtheta')
plt.subplot(5,1,5)
plt.plot(t_span, policy[4,:] ,'m-',linewidth=2)
plt.ylabel('u')
plt.show()
```
