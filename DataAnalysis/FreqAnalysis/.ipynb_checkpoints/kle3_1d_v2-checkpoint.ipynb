```python
import random
import math

import autograd.numpy        as np
import autograd.numpy.random as npr 

from autograd.scipy.stats    import norm
from autograd.scipy.linalg   import solve_continuous_are
from autograd                import jacobian, grad
import matplotlib.pyplot     as plt
```

```python
single_integrator = False # true for single integrator, false for double integrator
```

## Environment
1-D environments where a robot has access in the $x$-space.

```python
class SingleIntegrator_1DPointMassEnv(object): 
    def __init__(self): 
        self.dt = 0.1
        self.num_states = 1
        self.num_actions = 1
        self.explr_idx = [0]

        self._A = np.array([[0]])
        self._B = np.array([[1]]) 
    
    def get_lin(self,x=0,u=0): 
        return self._A, self._B

    def reset(self,state=None):
        if state is None: 
            self.state = npr.uniform(-0.1,0.1,size=(self.num_states,1))
        else: 
            self.state = state.copy()
        return self.state.copy()
    
    def f(self,x,u): # transiton model
        return self._A @ x + self._B @ u 
    
    def step(self,u): 
        self.state = self.state + self.f(self.state,u) * self.dt
        return self.state.copy()
```

```python
class DoubleIntegrator_1DPointMassEnv(object): 
    def __init__(self): 
        self.dt = 0.1
        self.num_states = 2
        self.num_actions = 1
        self.explr_idx = [0]

        self._A = np.array([[0,1],
                            [0,0]])
        self._B = np.array([[0],
                            [1]]) 
    
    def get_lin(self,x=0,u=0): 
        return self._A, self._B

    def reset(self,state=None):
        if state is None: 
            self.state = npr.uniform(-0.1,0.1,size=(self.num_states,1))
        else: 
            self.state = state.copy()
        return self.state.copy()
    
    def f(self,x,u): # transiton model
        return self._A @ x + self._B @ u 
    
    def step(self,u): 
        self.state = self.state + self.f(self.state,u) * self.dt
        return self.state.copy()
```

```python
class LQR(object):
    def __init__(self,model): 
        A,B =  model.get_lin()
        if model.num_states == 1:
            Q = np.diag([10])
            R = np.diag([10])
        elif model.num_states == 2: 
            Q = np.diag([10,1]) 
            R = np.diag([100])
        P = solve_continuous_are(A,B,Q,R)
        self.Klqr = np.linalg.inv(R) @ B.T @ P
    
    def dx(self): 
        return -self.Klqr.copy()
    
    def __call__(self,x): 
        return - self.Klqr @ x
```

```python
# testing LQR
if single_integrator: 
    env    = SingleIntegrator_1DPointMassEnv()
    policy = LQR(env)

    x0 = np.vstack([5])
else: 
    env    = DoubleIntegrator_1DPointMassEnv()
    policy = LQR(env)

    x0 = np.vstack([5,0])
    
x = env.reset(x0)
N = 100
x_list = np.zeros((N,env.num_states))
u_list = np.zeros((N,env.num_actions))
for i in range(N):
    mu = policy(x)
    x_list[i,:] = x.T
    u_list[i,:] = mu
    x = env.step(mu)


# plot
tspan = [i * env.dt for i in range(N)]
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(tspan,x_list)
plt.xlabel('t')
plt.legend(['x','x_dot'])
plt.grid()
plt.subplot(1,2,2)
plt.plot(tspan,u_list,'k')
plt.xlabel('t')
plt.legend(['$\mu$'])
plt.grid()
plt.suptitle('LQR control of point mass (single integrator)')
plt.show()
```

## Create a Buffer

```python
class ReplayBuffer:
    def __init__(self, capacity):
        self.capacity = capacity
        self.buffer = []
        self.position = 0
    
    def push(self, state):
        if len(self.buffer) < self.capacity:
            self.buffer.append(None)
        self.buffer[self.position] = state
        self.position = (self.position + 1) % self.capacity
    
    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        return np.stack(batch)

    def __len__(self):
        return len(self.buffer)
```

```python
class TargetDist(object): 
    def __init__(self):
        pass
    # target distribution functions
    def p_pdf(self,s): 
        sigmas = np.array([0.5,0.8])
        mus = np.array([-1.,3.])
        pis = np.array([0.4,0.6])
        out = 0
        for mu,var,pi in zip(mus,sigmas,pis):
            gaussian = 1/np.sqrt(2*np.pi*var**2)*np.exp(-1/2*(s-mu)**2/var**2) 
            out += pi * gaussian
        return out + 1e-4

tdist = TargetDist()
s = np.linspace(-5, 5, 100)
p = tdist.p_pdf(s)
# view distribution
plt.plot(s,p)
plt.title('target dist')
plt.xlabel('x')
plt.ylabel('likelihood')
plt.show()
```

```python
class kle3(object): 
    def __init__(self,env,horizon=100,capacity=100):
        # time parameters
        self.horizon     = horizon # N
        self.capacity    = 100
        self.buffer      = ReplayBuffer(capacity)
        self.u_list      = [np.zeros(env.num_actions) for _ in range(horizon)]
        self.dmu_star    = [np.zeros(env.num_actions) for _ in range(horizon)]
        self.num_states  = env.num_states
        self.num_actions = env.num_actions
        self.explr_dim   = env.explr_idx
    
    def psi(self,x,s): # (footprint in ian's original implementation)
        return np.exp(-0.5*np.square(s-x)/self.var)
#         return np.exp(-0.5*np.einsum('ij,ij->ij',(s-x) @ (self.SigmaI),(s-x)))
    
    def dpsi_dx(self,x,s): # (alt/eqivalent is jacobian(self.psi)), takes single x 
        inner = self.psi(x[self.explr_dim].T,s)*(s-x[self.explr_dim].T)/self.var
        return np.hstack([inner,np.zeros((s.shape[0],env.num_states-len(self.explr_dim)))])
#         return np.hstack([np.einsum('ij,ij->ij',self.psi(x[self.explr_dim],s)  @ (self.SigmaI),(s-x[self.explr_dim])),np.zeros((s.shape[0],env.num_states-len(self.explr_dim)))])
                      
    def q_pdf(self,x,s): # (pdf in ian's original implementation), takes single s
        return np.mean(self.psi(x[:,self.explr_dim],s)) + 1e-5
        
    # boundary
    def barrier(self, x):
        return 100.*(x[self.explr_dim]/10.)**8
    
    def dbarrier_dx(self,x): 
        return np.vstack([80.*(x[self.explr_dim]/10.)**7,np.zeros((env.num_states-len(self.explr_dim),1))])
    
    # controller
    def __call__(self,state,model,policy,target_dist,batch_size=20,Ns=40,clipat=2):
        # setup
        self.var = 0.2
        self.R = 0.02
        self.SigmaI = 1./self.var*np.eye(len(self.explr_dim))       # positive definite, width of guassian at each s (v=1) (called std in ian's original implmementation)
        self.RI     = 1./self.R*np.eye(self.num_actions)       # regularization term (called eps in ian's original implementations)
        
        derivs = []
        traj   = []
        x      = model.reset(state)
        s      = npr.uniform(-5,5,size=(Ns,len(self.explr_dim))) # column vector
        p      = target_dist.p_pdf(s)   # target distribution (column vector)
        p      = p.reshape(-1,1) 
        p     /= np.sum(p) # normalize

        # forward simulate
        for t in range(self.horizon):
            # compute derivatives
            A, B   = model.get_lin(x, self.u_list[t])
            dmudx  = policy.dx()
            gdx    = self.dpsi_dx(x,s)
            bdx    = self.dbarrier_dx(x)
            
            # store derivatives
            derivs.append((gdx,A,B,dmudx,bdx)) # note: ian's original implementation also had a barrier and dmudx=0
            traj.append(x)
            
            # step forward
            self.u_list[t] = policy(x)
            x = model.step(self.u_list[t])

        path = np.stack(traj)
        # add history
        if len(self.buffer) > batch_size:
            past_states = self.buffer.sample(batch_size)
            path = np.vstack([path,past_states])
        path = path.reshape(-1,self.num_states)

        # calculate spatial distribution
        q     = np.array([self.q_pdf(path,si) for si in s])
        q     = q.reshape(-1,1) 
        q    /= np.sum(q) # normalize

        # backwards integrate
        like_ratio = p/q # ratio of pdfs
        rho = np.zeros(x.shape) # set terminal condition = 0
        for t in reversed(range(self.horizon)): 
            dpsidx, dfdx, dfdu, dmudx,bdx = derivs[t] # dpsidx = gdx, dfdx = A, dfdu =  = h, dmudx = -K
            rho = rho - model.dt*( (like_ratio.T @ dpsidx).T/Ns - (dfdx + dfdu @ dmudx).T @ rho ) # - bdx*model.dt this is wehre you'd subtract the barrier gradient
            self.dmu_star[t] = -self.RI @ dfdu.T @ rho
            self.u_list[t] = np.clip( self.dmu_star[t] + self.u_list[t] ,-clipat,clipat)
        return self.dmu_star[0] 
```

## Instantiate the environment, model, target distribution, and KL-E^3

```python
npr.seed(0)

if single_integrator: 
    env      = SingleIntegrator_1DPointMassEnv()
    model    = SingleIntegrator_1DPointMassEnv()
else: 
    env      = DoubleIntegrator_1DPointMassEnv()
    model    = DoubleIntegrator_1DPointMassEnv()
policy   = LQR(env)
kl_explr = kle3(env,horizon=15)
tdist    = TargetDist()
```

## Simulate a trajectory 

```python
trajectory = []
u_list = []
mu_list = []
dmu_list = []
# x0 = np.array([[3]])
state = env.reset()
max_u = 5.
for _ in range(600):
    dmu = kl_explr(state,model,policy,tdist,clipat=max_u) 
    mu = policy(state)
    a = np.clip(dmu + mu,-max_u,max_u)
    trajectory.append(state)
    u_list.append(a)
    dmu_list.append(dmu)
    mu_list.append(mu)
    state = env.step(a)
    kl_explr.buffer.push(state)
```

## Plot trajectory and resulting distributions

```python
tf = 500
traj = np.hstack(trajectory).T
controls = np.hstack(u_list).T
policy_controls = np.hstack(mu_list).T
kle3_controls = np.hstack(dmu_list).T
tspan = [i * env.dt for i in range(tf)]

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(tspan,traj[:tf])
plt.xlabel('t')
plt.legend(['x','x_dot'])
plt.grid()
plt.subplot(1,2,2)
plt.plot(tspan,policy_controls[:tf],'k')
plt.plot(tspan,controls[:tf])
plt.plot(tspan,kle3_controls[:tf],'r')
plt.ylim([-max_u,max_u])
plt.xlabel('t')
plt.legend(['$\mu$','$\mu*$(clipped)','$\delta\mu*$'])
plt.grid()
plt.suptitle('Ergodic control of point mass (single integrator)')
plt.show()
```

```python
xx = np.linspace(-5, 5, 100)
scale = 500

target_dist = [tdist.p_pdf(xi) * scale for xi in xx]
target_dist = np.array(target_dist)
target_dist /= np.sum(target_dist)
target_dist *= scale

time_dist = [kl_explr.q_pdf(traj[:tf], xi) * scale for xi in xx]
time_dist = np.array(time_dist)
time_dist /= np.sum(time_dist)
time_dist *= scale

### Fourier decomp
num_basis = 20
cks = [np.mean(np.cos(k * (traj[:tf,0] + 5.0) * np.pi / (10.))) for k in range(num_basis)]
print(len(cks))
recon = np.zeros(xx.shape)
for k in range(num_basis):
    recon += cks[k] * np.cos(k * (xx + 5.0) * np.pi/10.)
recon /= np.sum(recon)
recon *= scale

## plot
fig, axs = plt.subplots(1,2, figsize=(8,4)) #, gridspec_kw={'height_ratios': [0.8, 1]})

axs[0].plot(tspan, traj[:tf,0], alpha=0.8, label='x(t)', color='tab:orange')
axs[1].plot(np.zeros(traj[:tf,0].shape), traj[:tf,0], 'o', alpha=0.01, color='m')
axs[1].plot(np.zeros(traj[50:52,0].shape), traj[50:52,0], 'o', alpha=0.2, color='m', label='samples')

axs[1].plot(target_dist,xx, 'tab:gray', alpha=0.5, lw=3, label='p(s)')
axs[1].plot(time_dist,xx, 'tab:blue', alpha=1.0, lw=2, label='q(s | x(t))')
axs[1].plot(recon,xx, 'm', alpha=0.5, lw=2, label='c(s | x(t))')

# formatting
axs[0].set_xlabel('t')
# axs[0].set_ylim(-5,5)
axs[1].set_ylabel('x')
axs[1].set_xlabel('likelihood')
# axs[1].set_ylim(-5,5)

axs[0].legend()
axs[1].legend( bbox_to_anchor=(1.05, 1), loc='upper left')

axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```
