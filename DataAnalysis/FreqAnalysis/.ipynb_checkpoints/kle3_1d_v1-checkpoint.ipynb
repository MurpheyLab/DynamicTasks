```python
import random
import math

import autograd.numpy        as np
import autograd.numpy.random as npr 

from autograd.scipy.stats import norm
from autograd             import jacobian, grad
import matplotlib.pyplot  as plt
```

## Environment
1-D environment where a robot has access in the $x$-space.

```python
class SingleIntegrator_1DPointMassEnv(object): 
    def __init__(self): 
        self.dt = 0.1
        self.num_states = 1
        self.num_actions = 1

        self._A = np.array([[0]])
        self._B = np.array([[1]]) 
    
    def get_lin(self,x=0,u=0): 
        return self._A, self._B

    def reset(self,state=None):
        if state is None: 
            self.state = npr.uniform(-0.1,0.1,size=(self.num_states,1))
        else: 
            self.state = state.copy()
        return self.state.copy()
    
    def f(self,x,u): # transiton model
        return self._A @ x + self._B @ u 
    
    def step(self,u): 
        self.state = self.state + self.f(self.state,u) * self.dt
        return self.state.copy()
```

## Create a Buffer

```python
class ReplayBuffer:
    def __init__(self, capacity):
        self.capacity = capacity
        self.buffer = []
        self.position = 0
    
    def push(self, state):
        if len(self.buffer) < self.capacity:
            self.buffer.append(None)
        self.buffer[self.position] = state
        self.position = (self.position + 1) % self.capacity
    
    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        return np.stack(batch)

    def __len__(self):
        return len(self.buffer)
```

```python
class LQR(object):
    def __init__(self,model): 
        self._model = model
        self._Q = np.diag([1])
        self._R = np.diag([10])
        self._RI = np.linalg.inv(self._R)
        self._P1 = np.diag([10])
        self.T = 0.2
        self.dt = 0.01
        self.tspan = np.arange(0,self.T,self.dt)
        self.N = self.tspan.shape[0]
        self.P = self.get_P()

    def J(self,x,u): # ref only
        cost = 0
        for (xt,ut) in zip(x,u): 
            cost += xt.T @ self._Q @ xt + ut.T @ self._R @ ut
        cost += x[-1].T @ self._P1 @ x[-1]
        cost /= 2
        return cost
    
    def dJ(self,x,u,z,v): # ref only
        cost = 0
        for (xt,ut) in zip(x,u): 
            cost += xt.T @ self._Q @ z + ut.T @ self._R @ v
        cost += x[-1].T @ self._P1 @ z
        return cost       
    
    def dP(self,P): 
        A,B =  self._model.get_lin()
        return -(P @ A + A.T @ P - P @ B @ self._RI @ B.T @ P + self._Q )
        
    def get_P(self):
        P = self._P1.copy()
        for t in range(self.N-1): 
            P = P - self.dP(P)*self.dt
        return P
            
    def K(self): 
        A,B =  self._model.get_lin()
        return self._RI @ B.T @ self.P 
    
    def __call__(self,x): 
        return - self.K() @ x
```

```python
# testing LQR
env    = SingleIntegrator_1DPointMassEnv()
policy = LQR(env)

x0 = np.array([3])
x = env.reset(x0)
N = 100
x_list = np.zeros((N,1))
u_list = np.zeros((N,1))
for i in range(N):
    mu = policy(x)
    x_list[i,:] = x 
    u_list[i,:] = mu
    x = env.step(mu)
    
# plot
tspan = [i * env.dt for i in range(N)]
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(tspan,x_list)
plt.xlabel('t')
plt.legend(['x'])
plt.grid()
plt.subplot(1,2,2)
plt.plot(tspan,u_list,'k')
plt.xlabel('t')
plt.legend(['$\mu$'])
plt.grid()
plt.suptitle('LQR control of point mass (single integrator)')
plt.show()
```

```python
class kle3(object): 
    def __init__(self,env,horizon=100,capacity=100):
        # time parameters
        self.horizon     = horizon # N
        self.capacity    = 100
        self.buffer      = ReplayBuffer(capacity)
        self.u_list      = [np.zeros(env.num_actions) for _ in range(horizon)]
        self.dmu_star    = [np.zeros(env.num_actions) for _ in range(horizon)]
        self.num_states  = env.num_states
        self.num_actions = env.num_actions
        
    # target distribution functions
    def p_pdf(self,s): 
        sigmas = np.array([0.5,0.8])
        mus = np.array([-1.,3.])
        pis = np.array([0.4,0.6])
        out = 0
        for mu,var,pi in zip(mus,sigmas,pis):
            gaussian = 1/np.sqrt(2*np.pi*var**2)*np.exp(-1/2*(s-mu)**2/var**2) 
            out += pi * gaussian
        return out + 1e-4
    
    # spatial distribution functions (eqn 3)
    def psi(self,x,s): # (footprint in ian's original implementation)
        return np.exp(-0.5*np.einsum('ij,ij->ij',(s-x) @ (self.SigmaI),(s-x)))
#         return np.exp(-0.5*np.sum(np.square(s-x)/self.Sigma,1)) # 1D only (ex sigma = 0.05)
    
    def dpsi_dx(self,x,s): # (alt/eqivalent is jacobian(self.psi)), takes single x 
        return np.einsum('ij,ij->ij',self.psi(x,s)  @ (self.SigmaI),(s-x))
#         return np.diag(self.psi(s,x)*(s-x)/ self.Sigma) # 1D only (ex sigma = 0.05)
                      
    def q_pdf(self,x,s): # (pdf in ian's original implementation), takes single s
        return np.mean(self.psi(x,s)) + 1e-5
        
    # controller
    def __call__(self,state,model,policy,batch_size=20,Ns=40,clipat=2):
        # setup
        self.SigmaI = 1.*np.eye(1)       # positive definite, width of guassian at each s (v=1) (called std in ian's original implmementation)
        self.RI     = 80.*np.eye(1)       # regularization term (called eps in ian's original implementations)
        
        derivs = []
        traj   = []
        x      = model.reset(state)
        s      = npr.uniform(-5,5,size=(Ns,self.num_states)) # column vector
        p      = self.p_pdf(s)   # target distribution (column vector)
        p     /= np.sum(p) # normalize

        # forward simulate
        for t in range(self.horizon):
            # compute derivatives
            A, B   = model.get_lin(x, self.u_list[t])
            dmudx  = -policy.K()
            gdx    = self.dpsi_dx(x,s)
            
            # store derivatives
            derivs.append((gdx,A,B,dmudx)) # note: ian's original implementation also had a barrier and dmudx=0
            traj.append(x)
            
            # step forward
            self.u_list[t] = policy(x)
            x = model.step(self.u_list[t])

        path = np.stack(traj)
        # add history
        if len(self.buffer) > batch_size:
            past_states = self.buffer.sample(batch_size)
            path = np.vstack([path,past_states])
        path = path.reshape(-1,self.num_states)

        # calculate spatial distribution
        q     = np.array([self.q_pdf(path,si) for si in s])
        q     = q.reshape(-1,1) 
        q    /= np.sum(q) # normalize
        
        # backwards integrate
        like_ratio = p/q # ratio of pdfs
        rho = np.zeros(x.shape) # set terminal condition = 0
        for t in reversed(range(self.horizon)): 
            dpsidx, dfdx, dfdu, dmudx = derivs[t] # dpsidx = gdx, dfdx = A, dfdu =  = h, dmudx = -K
            rho = rho - model.dt*( (like_ratio.T @ dpsidx)/Ns - (dfdx + dfdu @ dmudx).T @ rho ) # this is wehre you'd subtract the barrier gradient
            self.dmu_star[t] = -self.RI @ dfdu.T @ rho
            self.u_list[t] = np.clip( self.dmu_star[t] + self.u_list[t] ,-clipat,clipat)
        return self.dmu_star[0] 
```

## Instantiate the environment, model, target distribution, and KL-E^3

```python
npr.seed(0)

env      = SingleIntegrator_1DPointMassEnv()
model    = SingleIntegrator_1DPointMassEnv()
policy   = LQR(env)
kl_explr = kle3(env,horizon=15)
```

## Simulate a trajectory 

```python
trajectory = []
u_list = []
mu_list = []
dmu_list = []
# x0 = np.array([[3]])
state = env.reset()
for _ in range(600):
    dmu = kl_explr(state,model,policy,clipat=2) 
    mu = policy(state)
    a = np.clip(dmu + mu,-2,2)
    trajectory.append(state)
    u_list.append(a)
    dmu_list.append(dmu)
    mu_list.append(mu)
    state = env.step(a)
    kl_explr.buffer.push(state)
```

## Plot trajectory and resulting distributions

```python
tf = 400
traj = np.hstack(trajectory).T
controls = np.hstack(u_list).T
policy_controls = np.hstack(mu_list).T
kle3_controls = np.hstack(dmu_list).T
tspan = [i * env.dt for i in range(tf)]

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(tspan,traj[:tf])
plt.xlabel('t')
plt.legend(['x'])
plt.grid()
plt.subplot(1,2,2)
plt.plot(tspan,policy_controls[:tf],'k')
plt.plot(tspan,controls[:tf])
plt.plot(tspan,kle3_controls[:tf],'r')
plt.ylim([-2,2])
plt.xlabel('t')
plt.legend(['$\mu$','$\mu*$(clipped)','$\delta\mu*$'])
plt.grid()
plt.suptitle('Ergodic control of point mass (single integrator)')
plt.show()
```

```python
xx = np.linspace(-5, 5, 100)
scale = 500

target_dist = [kl_explr.p_pdf(xi) * scale for xi in xx]
target_dist = np.array(target_dist)
target_dist /= np.sum(target_dist)
target_dist *= scale

time_dist = [kl_explr.q_pdf(traj[:tf], xi) * scale for xi in xx]
time_dist = np.array(time_dist)
time_dist /= np.sum(time_dist)
time_dist *= scale

### Fourier decomp
num_basis = 20
cks = [np.mean(np.cos(k * (traj[:tf,0] + 5.0) * np.pi / (10.))) for k in range(num_basis)]
recon = np.zeros(xx.shape)
for k in range(num_basis):
    recon += cks[k] * np.cos(k * (xx + 5.0) * np.pi/10.)
recon /= np.sum(recon)
recon *= scale

## plot
fig, axs = plt.subplots(1,2, figsize=(6,4)) #, gridspec_kw={'height_ratios': [0.8, 1]})

axs[0].plot(tspan, traj[:tf,0], alpha=0.8, label='x(t)', color='tab:orange')
axs[1].plot(np.zeros(traj[:tf,0].shape), traj[:tf,0], 'o', alpha=0.01, color='m')
axs[1].plot(np.zeros(traj[50:52,0].shape), traj[50:52,0], 'o', alpha=0.2, color='m', label='samples')

axs[1].plot(target_dist,xx, 'tab:gray', alpha=0.5, lw=3, label='p(s)')
axs[1].plot(time_dist,xx, 'tab:blue', alpha=1.0, lw=2, label='q(s | x(t))')
axs[1].plot(recon,xx, 'm', alpha=0.5, lw=2, label='c(s | x(t))')

# formatting
axs[0].set_xlabel('t')
axs[0].set_ylim(-5,5)
axs[1].set_ylabel('x')
axs[1].set_xlabel('likelihood')
axs[1].set_ylim(-5,5)

axs[0].legend()
axs[1].legend( bbox_to_anchor=(1.05, 1), loc='upper left')

axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```
